```python
## 데이터프레임을 넣고 column별 특성 및 결측값, 고유값들을 확인하는 함수를 작성해본다.
## 필수는 아니지만 전체적인 흐름을 파악하기 쉬워진다.

def resumetable(df):
  print(f'데이터셋 크기: {df.shape}')                                # 데이터프레임의 전체 크기(행, 열) 출력

  summary = pd.DataFrame(df.dtypes, columns=['데이터 타입'])         # 각 피처의 데이터 타입을 가져와 데이터프레임으로 생성
  summary = summary.reset_index()                                    # 인덱스를 초기화하여 컬럼으로 변환
  summary = summary.rename(columns={'index':'피처'})                 # 'index' 컬럼명을 '피처'로 변경

  summary['결측값 개수'] = df.isnull().sum().values                 # 각 피처의 결측값(null) 개수 계산
  summary['고유값 개수'] = df.nunique().values                      # 각 피처의 고유값 개수 계산

  summary['첫 번째 값'] = df.loc[0].values                          # 각 피처의 첫 번째 샘플 값
  summary['두 번째 값'] = df.loc[1].values                          # 각 피처의 두 번째 샘플 값
  summary['세 번째 값'] = df.loc[2].values                          # 각 피처의 세 번째 샘플 값

  return summary                                                     # 요약 테이블 반환
```

### 카이제곱검정 연습
- 가입개월수와 이탈률은 연관이 있는가?

```python
# 교차표의 형태로 만들기
monthchurn_crosstab = pd.crosstab(tel['가입개월수'], tel['이탈여부'])
monthchurn_crosstab
```

```python
# 카이제곱 독립성 검정 실행
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(monthchurn_crosstab)
```

- chi2: 검정 통계량
- p: p-value
- dof: 자유도
- expected: 기대빈도표

```python
# 결과 해석
print(f"Chi-squared statistic: {chi2}")
print(f"p-value: {p}")

if p < 0.05:
    print("→ 유의미한 관계 있음 (귀무가설 기각)")
else:
    print("→ 유의미한 관계 없음 (귀무가설 채택)")
```

### 0개월 탈락시키고, 카이제곱 검정 진행

```python
monthchurn_crosstab = monthchurn_crosstab.iloc[1:]

chi2, p, dof, expected = chi2_contingency(monthchurn_crosstab)
```

```python
# 결과 해석
print(f"Chi-squared statistic: {chi2}")
print(f"p-value: {p}")

if p < 0.05:
    print("→ 유의미한 관계 있음 (귀무가설 기각)")
else:
    print("→ 유의미한 관계 없음 (귀무가설 채택)")
```

```python
plt.figure(figsize=(12, 6))

# 각 이탈 여부(True, False)에 대해 선 그리기
for churn_status in monthchurn_crosstab.columns:
    plt.plot(monthchurn_crosstab.index, monthchurn_crosstab[churn_status], label=str(churn_status))

plt.title('가입개월수에 따른 이탈여부 분포')
plt.xlabel('가입개월수')
plt.ylabel('가입자 수')
plt.legend(title='이탈여부')
plt.grid(True)
plt.tight_layout()
plt.show()
```
